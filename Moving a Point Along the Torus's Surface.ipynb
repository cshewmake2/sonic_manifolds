```python
%matplotlib notebook
from ipywidgets import *
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

import torus
```

```python
R = 5
r = 1

ths = 2*np.pi*np.random.rand(1,1000)
phs = 2*np.pi*np.random.rand(1,1000)
pts = np.array([torus.torus(th,ph,R,r) for th in ths for ph in phs])

# Plot the torus
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(pts[:,0],pts[:,1],pts[:,2])
plt.xlim(-8, 8)
plt.ylim(-8, 8)
ax.set_zlim(-4,4)
plt.autoscale(False)


def update(th = 2*np.pi, ph = 2*np.pi, R = 5.0, r = 1.0):
    plt.cla()
    # plot torus
    pts = np.array([torus.torus(th,ph,R,r) for th in ths for ph in phs])
    ax.scatter(pts[:,0],pts[:,1],pts[:,2])
    plt.xlim(-10, 10)
    plt.ylim(-10, 10)
    ax.set_zlim(-4,4)
    plt.autoscale(False)
    
    # plot point
    x,y,z = torus.torus(th,ph,R,r)
    plt.plot([x], [y], [z], marker='o', markersize=20, color="red")
    plt.show()

interact(update);
```
